```python
# IMPORTING LIBRARY.....
import matplotlib.pyplot as plt
import numpy  as NP
import cv2 as CV
```

* FOLLOWING SNIPPET OF CODE IS FOR CREATING FUNCTION NAMED AS CONV WHICH WILL PERFORM CONVOLUTION OPERATION.
* APPROACH:
* IN LINE(1) IT IS TAKING TWO ARGUMENT ONE IS AN IMAGE AND SECOND ONE IS AN SOBEL EDGE DETECTOR OPERATOR OF SIZE 3x3
* FROM LINE(2) to LINE(10) WE HAVE PAD ZEROS OUTSIDE OUR IMAGE SO THAN 

```python
def conv(IM,K):
    #......CREATING MATRIX OF ALL ZEROS FOR ZERO PADDIN
    Z_H = NP.zeros((IM.shape[0]+2,IM.shape[1]+2))
    Z_V = NP.zeros((IM.shape[0]+2,IM.shape[1]+2))
    F = NP.zeros((IM.shape[0]+2,IM.shape[1]+2))
    
    
    #......PADDING ZEROS
    Z_H[1:Z_H.shape[0]-1,1:Z_H.shape[1]-1] = IM
    Z_V[1:Z_V.shape[0]-1,1:Z_V.shape[1]-1] = IM
    
    # FOLLOWING IS THE LOOP FOR PERFROMING CONVOLUTION OF 3X3 SOBEL OPERTOR KERNEL WITH ALL 3X3 CELL OF AN IMAGE
    
    for i in range(0,Z_H.shape[0]-2):
        for j in range(0,Z_H.shape[1]-2):
            Z_H[i,j]=NP.abs(sum(sum(Z_H[i:i+3,j:j+3]*K)))  #....HORIZANTAL EDGE IMAGE
            Z_V[i,j]=NP.abs(sum(sum(Z_V[i:i+3,j:j+3]*K.T))) #....VERTICAL EDGE IMAGE
            F[i,j] = Z_H[i,j]**2 + Z_V[i,j]**2 
            
    print(min(list(Z_H.ravel())))
    print(max(list(Z_H.ravel())))
    print(min(list(Z_V.ravel())))
    print(max(list(Z_V.ravel())))
    F = NP.sqrt(F) #.....MAGNITUDE EDGE OF GIVEN IMAGE
    
    #....DISPLAYING RESULT
    
    
    plt.figure(1)
    plt.imshow(Z_H,'gray')
    plt.title('HORIZONTAL EDGE')
    plt.axis('off')
    plt.figure(2)
    plt.imshow(Z_V,'gray')
    plt.title('VERTICAL EDGE')
    plt.axis('off')
    plt.figure(3)
    plt.imshow(F,'gray')
    plt.title('EDGE DETECTION BY TAKING MAGNITUDE')
    plt.axis('off')
    
```

```python
x = CV.imread("image2.jpg")
plt.figure(1)
plt.imshow(x)
B = x[:,:,0]
G = x[:,:,1]
R = x[:,:,2]
A = (B/3+G/3+R/3)
plt.imshow(A,'gray')
plt.title('ORIGINAL IMAGE')
plt.axis('off')
S = NP.array([[-1,-2,-1],[0,0,0],[1,2,1]])



```

* CALLING FUNCTION and PASSING TWO ARGUMENT IN IT ONE IS IMAGE(A) and second one is SOBEL operator

```python
conv(A,S) 
```
