```python
#....QUESTION 4:
import cv2 as CV
import numpy as NP
import matplotlib.pyplot as PL
```

* IT IS FUNCTION  FOR  PLOTING HISTOGRAM OF ANY GRAYSCALE IMAGE,ONLY ARGUMENT WE NEED TO PASS IN IT IS IMAGE

```python
def histogram(IM):
    temp = [0]*256
    for j in range(0,IM.shape[0]):
        for k in range(0,IM.shape[1]):
            count = IM[j,k]
            temp[int(count)] = temp[int(count)] + 1
    PL.stem(temp)
    return temp
                
```

```python
def RGB2GRAY(y):
    if x.shape[2]!=3 :
        I = y
        PL.figure()
        PL.imshow(I)
        PL.axis('off')
        PL.title('ORIGINAL GRAY IMAGE')
    if x.shape[2]==3:
        PL.figure(1)
        PL.imshow(y)
        PL.axis('off')
        PL.title('ORIGINAL COLOR IMAGE')
        B = y[:,:,0]
        G = y[:,:,1]
        R = y[:,:,2]
        A = (B+G+R)/3
        PL.figure(2)
        PL.imshow(A,'gray')
        PL.axis('off')
        PL.title('GRAYSCALE IMAGE')
    
    return A
x = CV.imread('image4.jpg')
I = RGB2GRAY(x)



```

```python
#......cALLING FUNCTION........
T = histogram(I)
```

* CALCULATING PROBABILITY AND CDF OF INTENSITY VALUE.........

```python
PROB = [0]*256
CDF =  [0]*256
for i in range(0,256):
    PROB[i] = T[i]/sum(T)
    CDF[i] = sum(PROB) 
```

* MAPPING THAT TRANSFORMED INTENSITY TO AN INPUT IMAGE

```python
for i in range(0,256):
    R = NP.where(I == i)
    for j in range(0,int(len(R[0]))):
        I[R[0][j],R[1][j]] = CDF[i]*255
```

```python

PL.imshow(I,'gray')
PL.title('IMAGE AFTER HISTOGRAM EQUALIZATION')
PL.axis('off')
```

* Histogram of TRASFORMED IMAGE
